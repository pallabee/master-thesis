```python
import matplotlib.pyplot as plt
import numpy as np

feature_size = np.arange(1, 4)
#GNN Validation accuracy and feature reconstruction loss for features generated using Diffusion Models for DBLP Author class 0 
validation_acc = [0.64, 0.75, 0.81]
feature_loss = [-568, -2575, -8252]

# Create a plot with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))
my_xticks = ['8', '50', '334']
plt.xticks(feature_size, my_xticks,fontsize=20)
# Plot on the first y-axis
ax1.plot(feature_size, validation_acc, label='validation_acc', color='blue', marker='o')
ax1.set_xlabel('feature_count',fontsize=20)
ax1.set_ylabel('validation_accuracy', color='blue',fontsize=20)
ax1.tick_params(axis='y', labelcolor='blue',labelsize=20)

# Create a second y-axis
ax2 = ax1.twinx()
ax2.plot(feature_size, feature_loss, label='feature_reconstruction_loss', color='green', marker='s')
ax2.set_ylabel('feature_reconstruction_loss', color='green', fontsize=20)
ax2.tick_params(axis='y', labelcolor='green',labelsize=20)


# Add labels and title
plt.title('Comparison of Validation Loss and Feature Reconstruction loss Over Feature Count')

# Show the plot
plt.show()
```
