```python
import networkx as nx
import matplotlib.pyplot as plt
import torch
```

```python
def plot_graph(G,attribute):

    color_class_map = {0: 'blue', 1: 'red', 2: 'darkgreen'}

    nx.draw(G, 
        with_labels=False, node_color=[color_class_map[node[1][attribute]] 
                        for node in G.nodes(data=True)], 
            node_size=100,
        font_color='white')
    plt.show() 
```

Real Graph

```python
G15_real = nx.read_gexf('real_G15.gexf')
plot_graph(G15_real,'node_type')
```

Synthetic Graphs

```python
graph1 = nx.read_gexf('sync908574194857/sync908574194857_G13.gexf')
```

```python
syn_class_path = 'sync908574194857/sync908574194857_node_class13.pt'
node_class_syn = torch.load(syn_class_path).detach().numpy()
nx.set_node_attributes(graph1, dict(zip(graph1.nodes(), node_class_syn)), 'class')
plot_graph(graph1,'class')
```
