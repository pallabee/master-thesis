```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import itertools
import community as community_louvain
import matplotlib.cm as cm
```

```python
path = '../approach_2/real_graphs/dblp/'
```

```python
G20 = nx.read_gexf(path + 'real_G20.gexf')
```

```python
def plot_graph(G):
    nx.draw(G, with_labels = True)
    plt.show()
```

```python
def plot_partition(G,partition):
    # draw the graph
    pos = nx.spring_layout(G)
    # color the nodes according to their partition
    cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
    nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=100,label=True,
                            cmap=cmap, node_color=list(partition.values()))
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    plt.show()
```

```python
def detect_community(G):

    partition = community_louvain.best_partition(G)
    plot_partition(G,partition)
```

```python
plot_graph(G20)
```

```python
detect_community(G20)
```

```python
list_com= nx.community.louvain_communities(G20, seed=123)
print(list_com)
H = G20.subgraph(list_com[0])
nx.draw(H, node_color='darkslateblue', node_size=100, with_labels = False)
```
